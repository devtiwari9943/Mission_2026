```python
## Accessing list elements 
fruits=["apple","banana","kiwi","cherry","gauva"]
print(fruits[0])
```

```python
print(fruits[0])#for any one perticular
print(fruits[-2])#total - no. = position 
print(fruits[0:4])#starting from index 0 ending at 4 (that is excluded )
print(fruits[0:])#all included 
print(fruits[-1:])#last
```

```python
#Modifying the list elements 
fruits[2]="watermelon"
print(fruits)
```

```python
#List methods 


#1) append-> to add an element in the end 
fruits.append("orange")
print(fruits)
```

```python
#2) insert -> insert at any perticular index 
fruits.insert(1,"tomato")
print(fruits)
```

```python
#3) remove-> to remove any element (remove the first occurance of an item )
fruits.remove("banana")
print(fruits)
```

```python
#4) pop ->remove and return the last element 
popped_fruits=fruits.pop()
print(popped_fruits)
print(fruits)
```

```python
#5) index-> to get the index of any item 
index=fruits.index("cherry")
print(index)
```

```python
#6)count -> count the no. of times any item comes 
fruits.insert(2,"gauva")
print(fruits.count("gauva"))
```

```python
#7) sort -> sort the  list  in ascending order
fruits.sort()
```
