```python
## Ex1 Temperature convertion

def temp_conversion(temp,unit):
    """This function converts temperature between  clesius to fahrenheit"""
    if unit=="C":
        f= (temp*1.8)+32## clesius to fahrenheit 
        return f
    elif unit =="F":
        c=(temp-32)*(5/9)## Fahrenheit to celsius
        return c


print(temp_conversion(25,"C"))
print(temp_conversion(77,"F"))
```

```python
## Ex2 Password strength checker 

def is_strong_password(password):
    """Check if the password is strong or not """
    if len(password)<8:
        return False 
    if not any(char.isdigit() for char in password ):
        return False
    if not any(char.islower() for char in password ):
        return False
    if not any(char.isupper() for char in password ):
        return False
    if not any(char in '!@#$%^&*()' for char in password ):
        return False
    return True


## calling the function 

print(is_strong_password("epfeif3r"))
print(is_strong_password("Apfei3*f3r"))
```

```python
# Ex 3 Calculate the total cost of item in ashoping cart 


def cal_total_cost(cart):
    total_cost=0
    for item in cart :
        total_cost+=item["price"]*item["qunatity"]
    return total_cost


## cart data 
cart=[
    {"name":"apple","price":5,"qunatity":8},
    {"name":"banana","price":6,"qunatity":9},
    {"name":"milk","price":36,"qunatity":2}

]

## calling function 

total_cost=cal_total_cost(cart)
print(total_cost)

```

```python
## Ex 4 check if a string is palindrome

def is_palindrome(text):
    text=text.lower().replace(" ","")
    return text == text[::-1]
    



print(is_palindrome("A man a plan a canal Panama"))
print(is_palindrome("hello"))
```

```python
## Ex 5 Factorial using recursion 3

def fact(num):
    if num==1:
        return 1 
    else:
        return num*fact(num-1)

print(fact(5))
```

```python
# Ex 6 A function to read a file and count the frequency of each word 

def count_word_frequency(file_path):
    word_count = {}

    with open(file_path, "r") as file:
        for line in file:
            words = line.split()

            for word in words:
                word = word.lower().strip(".,!?;:\"'")

                word_count[word] = word_count.get(word, 0) + 1

    return word_count


filepath = "sample.txt"
word_frequency = count_word_frequency(filepath)
print(word_frequency)




## skipped for now because i have not know about file reading 
```

```python
## Ex 7 validate email address 
import re

def is_valid_email(email):
    pattern = r'^[a-zA-Z0-9._+-]+@[a-zA-Z0-9-]+\.[a-zA-Z0-9-.]+$'
    return re.match(pattern, email) is not None

print(is_valid_email("test@example.com"))   # True
print(is_valid_email("invalid-email"))      # False




## skiped for now 
```
