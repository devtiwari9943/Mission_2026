```python
#sol 11 Create a tuple with the characters of a string. Join the tuple elements into a single string. Print the string.


# Store a string in the variable 'string'
string = "hello"

# Convert the string into a tuple.
# Since a string is iterable, each character becomes a separate element.
# Result: ('h', 'e', 'l', 'l', 'o')
tpl = tuple(string)

# Join all the characters of the tuple back into a single string.
# '' (empty string) means "don't add anything between the characters".
# If we used '-' instead of '', the result would be: h-e-l-l-o
joined_string = ''.join(tpl)

# Print the final joined string.
# Output: hello
print(joined_string)
```

```python
# sol 15  Write functions that take a tuple and return the minimum, maximum, and sum of the elements. Print the results for a sample tuple.

def min_in_tuple(tpl):
    return min(tpl)

def max_in_tuple(tpl):
    return max(tpl)

def sum_of_tuple(tpl):
    return sum(tpl)

sample_tpl = (1, 2, 3, 4, 5)
print(f"Minimum: {min_in_tuple(sample_tpl)}")
print(f"Maximum: {max_in_tuple(sample_tpl)}")
print(f"Sum: {sum_of_tuple(sample_tpl)}")
```
