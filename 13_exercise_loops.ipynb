```python
## ques- calculate the sum of first n natural numbers using a while loop and for loop 

# while loop 
a = int(input("enter any digit"))
sum = 0
i=1
while i!=a+1:
    sum =sum +i
    i=i+1
print(sum)

```

```python
# for loop 
a = int(input("enter any digit"))

sum=0
for i in range(a+1):
    sum=sum+i
print(sum)
```
