## Sets
Sets are a built in data type in python used to store collection of unique items. They are unordered, meaning that the element do not follow a specific order, and they do not allow duplicate elements. Sets are useful for membership tests, eliminiting duplicate entries, and performing mathematical set operaion like union, intersection, difference, and symmetric difference.

```python
## create a set 

my_set= {1,2,3,4,5,6,}
print(my_set)
print(type(my_set))
```

```python
my_empty_set=set()
print(type(my_empty_set))
```

```python
list=[1,2,3,4,5,6,7,5,4,3,2,6,7,5,4,]
new_list_to_set=set(list)
print(new_list_to_set)
```

## Basic set operation

```python
##  1) adding an element

new_set={1,2,3,4,5,6}
my_set.add(7)
print(my_set)
```

```python
## 2) removing element 
new_set.remove(3)
print(new_set)
```

```python
## 3)if element is not present in the set while removing it , it does not show any errpr

new_set.discard(11)
print(new_set)
```

```python
## 4) pop method  -> for popingout the first element


set1={1,2,3,4,5}
remove=set1.pop()
print(remove)
```

```python
## 5) clear all the elements

set1.clear()
print(set1)
```

```python
## set membership test 

set2={1,2,3,4,5,6}
print(3 in set2)
print(10 in set2)
```

```python
# mathematical operation

set3={1,2,3,4,5,6}
set4={5,6,7,8,9, 10}


## union 

union_set=set3.union(set4)
print(union_set)


#intersection

intersection_set=set3.intersection(set4)
print(intersection_set)



# if we want to intersection and update the set al together 

set3.intersection_update(set4)
print(set3)
```

```python
## difference 


set1={1,2,3,4,5,6,7,} 
set2={4,5,6,7,8}
print(set1.difference(set2))# that means set1 - set2
```

```python
## symmetric difference 

set1.symmetric_difference(set2)# that means combine both the set and removed tha elements present in both the set 

```

```python
## Set methods 


set1={1,2,3,4,5}
set2={3,4,5}


# is subset
print(set1.issubset(set2))# checking

print(set1.issuperset(set2))# cjhecking again 

```

```python
lt=[1,2,3,4,5,3,4,2,]
set_lst=set(lt)
print(set_lst)
```

```python
##counting unique words in text 

text=' In this tutorial we are discussing about sets'
words=text.split()

## convert list of worde to set to get uique words 
unique_words=set(words)
print(unique_words)
print(len(unique_words))
```
