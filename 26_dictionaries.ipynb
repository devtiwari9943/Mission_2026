## Dictionaries 
Dictionaries are unordered collection of items. They store data in key-value pairs.
 Keys must be unique and immutable (eg., string, numbers, or tuple), while values can be of any type


```python
## Creating dictionaries 
empty_dict={}
print(type(empty_dict))
```

```python
empty_dict=dict()
print(type(empty_dict))
```

```python
student={"name":"Dev","age":"22","grade":24}
print(student)
print(type(student))
```

```python
# single key is always used (it kinda a type of error )
student={"name":"Dev","age":"22","name":24}
print(student)
print(type(student))
```

```python
# accessing dictionaries element

student={"name":"Dev","age":"22","grade":"A"}
print(student)


print(student['grade'])
print(student['name'])
```

```python
## Accessing using get() method



print(student.get('grade'))
print(student.get('name'))
print(student.get('last_name'))# if the key is not present it will show none in output

# if the key is not present and we want a default outupt for that 
print(student.get('last_name',"Not available "))
```

```python
## Modifying dictionaries elements 
## dictionaries are mutable, so you can update , add, delete elements

print(student)
```

```python
student["age"]=33##update
print(student)
student['Address']="India"  ## added a new key and value
print(student)
```

```python
del student['grade']## delete key and value 
print(student)
```

```python
## Dictionaries methods 

keys=student.keys()## get all the keys
print(keys)

values=student.values()## get all the values
print(values)


items=student.items()## get all the key value pairs 
print(items)


```

```python
## Shallow copy


student_copy=student
print(student)
print(student_copy)
```

```python
student["name"]="prabhu"## here the name will be updated to both the dictionaries 
print(student)
print(student_copy)
```

```python
## but if we want that the copied dictionaries shoud not updated while updating the original dict

student_copy1=student.copy()## shallow copy 
print(student_copy1)
print(student)

print("after updation using copy function")
student["name"]="naimish"
print(student_copy1)
print(student)
```

```python
## Iteration over dictionaries 
## you can use loops to iterate over dictionaries , keys , values , items 



## iterating over keys
for keys in student.keys():
    print(keys)
```

```python
## Iterate over values 
for values in student.values():
    print(values)
```

```python
## iterate over key values pairs 

for key, values in student.items():
    print(f"{keys}:{values}")
```

```python
## nested dictionaries 

students={
    "Student1":{"name":"Dev","age":23},
    "Student2":{"name":"harshit","age":20},
    "Student3":{"name":"yashraj","age":19}

}
print(students)
```

```python
## Accessing nested dictionaries elements 
print(students["Student2"]["name"])
print(students["Student2"]["age"])
```

```python
## Iterating over nested dictionaries 

for student_id, student_info in students.items():
    print(f"{student_id}:{student_info}")
    for key,values in student_info.items():
        print(f"{key}:{values}")
```

```python
## Dictionaries comprehension 

squares={x:x**2 for x in range(5)}
print(squares)
```

```python
## Condetion dictionaries comprehension 

evens={x:x**2 for x in range(11) if x%2==0}
print(evens)
```

```python
## Practical examples 

## use a dictonary to count the frequency of element in a list 


num=[1,2,3,4,3,2,1,5,6,8,9,7,5,3,4,5,7,6,4,2,2,1,2,4,5,6,7,8,9,5,]
frequency={}

for number in num:
    if number in frequency:
        frequency[number]+=1
    else:
            frequency[number]=1
print(frequency)
```

```python
## Merge 2 dict into one 


dict1={"a":1,"b":2}
dict2={"c":3,"d":4}
merged_dict={**dict1,**dict2}
print(merged_dict)
```
