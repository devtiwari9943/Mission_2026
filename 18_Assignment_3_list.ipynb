```python
#sol 9   (currently skiped )

def transpose_matrix(matrix):
    transposed = [[matrix[j][i] for j in range(len(matrix))] for i in range(len(matrix[0]))]
    return transposed

matrix = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 9]
]
transposed = transpose_matrix(matrix)
print("Original matrix:")
for row in matrix:
    print(row)
print("Transposed matrix:")
for row in transposed:
    print(row)
```

```python
#sol 10 

def flat_list (nested_list):
    sin_list=[item for sub_list in nested_list for item in sub_list]
    return sin_list


nested_list = [
    [1,2,3],
    [4,5,6],
    [7,8,9]
]

flattened = flat_list(nested_list)
print("Original nested list:")
print(nested_list)
print("Flattened list:")
print(flattened)

```

```python
#sol13
def reversed_lst(lst):
    return lst[::-1]

lst=[1,2,3,4,5,6,]
rev_l=reversed_lst(lst)
print(f"Original list {lst}")
print(f"Riversed list {rev_l}")
```

```python
#sol14
def rotate_lst(lst,n):
    return lst[n:]  + lst[:n]
lst=[1,2,3,4,5,6,]
n=2
rot_lst=rotate_lst(lst,n)
print(f"Original list {lst}")
print(f"Rotated list{rot_lst}")
```

```python
#sol 15
def intersection_lst(list1,list2):
    return [x for x in list1 if x in list2]

list1=[1,2,3,4,5,8,9]
list2=[8,9,0,34,5]
int_lst=intersection_lst(list1,list2)
print(f"List1{list1}")
print(f"List2{list2}")
print(f"Intersected  list{int_lst}")
```
